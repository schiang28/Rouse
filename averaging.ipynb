```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('features.csv')
```

```python
df['BALANCE'] = (df['BALANCE'] - df['BALANCE'].min()) / (df['BALANCE'].max() - df['BALANCE'].min())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(df[['WON']])
df['WON'] = scaled
rankdiff = scaler.fit_transform(df[['RANKDIFF']])
df['RANKDIFF'] = rankdiff
df
```

```python
player_ids = {}
ID=1

for n in list(df['Name']):
    if n.lower() in player_ids:
        continue
    else:
        player_ids[n.lower()]=ID
        ID+=1

player_id = []
for n in list(df['Name']):
    player_id.append(player_ids[n.lower()])
```

```python
df['Player ID'] = player_id
```

```python
#df = df.drop(columns=['Name'])
df.to_csv('playerid.csv', index=False)
```

```python
player_averages=[]
for player in range(1, len(player_ids)+1):
    games = (df.loc[df['Player ID'] == player]).values.tolist()
    if player==7:
        print(games)
    averages = [0]*len(games[0])
    for i in range(len(games)):
        averages = [sum(j) for j in zip(averages, games[i])]
    averages = [i/len(games) for i in averages]
    #print(averages)
    player_averages.append(averages)
```

```python
columns = ['SP','RP','SRP','LRP','FHP','BHP','RANK','RANKDIFF','SA','SRA','FHA','BALANCE','Player ID']
profiles = pd.DataFrame(np.array(player_averages),
                   columns=columns)
```
