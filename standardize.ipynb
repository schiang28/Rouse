```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('features.csv').drop(columns=['Name'])
df
```

```python
df['BALANCE'] = (df['BALANCE'] - df['BALANCE'].min()) / (df['BALANCE'].max() - df['BALANCE'].min())
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(df[['WON']])
```

```python
df['WON'] = scaled
```

```python
rankdiff = scaler.fit_transform(df[['RANKDIFF']])
df['RANKDIFF'] = rankdiff
```

```python
df
```

```python
df.to_csv('data_scaled.csv', index=False)
```
